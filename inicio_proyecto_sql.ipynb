```python
import pandas as pd
df = pd.read_csv("cars.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

# Crear una instancia del codificador
le = LabelEncoder()

# Ajustar y transformar la columna Manufacturer
df['Manufacturer_encoded'] = le.fit_transform(df['Manufacturer'])

# Verificar los primeros registros para asegurarnos de que la conversión fue correcta
print(df[['Manufacturer', 'Manufacturer_encoded']].head())
```

```python
correlation = df[['Price', 'Manufacturer_encoded']].corr().iloc[0, 1]
print(f'Correlación entre Precio y Fabricante: {correlation}')
```

```python
import pandas as pd
import pymysql                        # for getting data from a SQL database
from sqlalchemy import create_engine  # for establishing the connection and authentication
from getpass import getpass
```

```python
password = getpass() #Contraseña del servidor MySQL
connection_string = 'mysql+pymysql://root:'+password+'@localhost/Proyecto'
engine = create_engine(connection_string)
query = '''SELECT t.Fuel_type, ROUND(AVG(c.Price),0) AS "Average Price", ROUND(AVG(c.Mileage),0)
FROM cars AS c
JOIN technique as t
ON c.ID = t.ID
GROUP BY t.Fuel_type
ORDER BY AVG(c.Price) DESC;'''
data = pd.read_sql_query(query, engine) #Traer datos
```
